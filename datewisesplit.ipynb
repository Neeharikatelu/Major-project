```python
import os
import pandas as pd
from os.path import exists

#Base directory path
base = r"/home/prio/loc_project/Mini_2_ERC_721/timestamp_out"

#Output directory path
output = r"/home/prio/loc_project/Mini_2_ERC_721/csv_daywise"

directory = base

#Iterate over all files in directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    print(f)
    
    try:
        # Read csv file from dataframe
        df = pd.read_csv(f)

    
        #Get unique timestamps from date columns
        dates = df['diffTimeStamp'].astype(str).str[:12].unique()
        
        #Iterate over all dates
        for d in dates:
            formatted_date = d.replace(" ", "_").replace(":", "-")
            fname = formatted_date + ".csv"
            
            #Slice the data of particular date
            df2 = df[df['diffTimeStamp'].astype(str).str[:12] == d]
            new_directory = output
            f_path = os.path.join(new_directory, fname)
            file_exists = exists(f_path)
            
            #Save to file if not exists
            if file_exists:
                df2.to_csv(f_path, mode='a', index=False, header=False)
            
            #Or append to already existing file
            else:
                df2.to_csv(f_path, index=False)
            print("written to file: ", f_path)
    except FileNotFoundError:
        print("File not found:", f)
    except PermissionError:
        print("Permission denied to write to output directory:", output)
    except Exception as e:
        print("An error occurred:", e)
```

```python
import pandas as pd
import os

input_directory = r"/home/prio/loc_project/Mini_2_ERC_721/timestamp_out"
output_directory_daywise = r"/home/prio/loc_project/Mini_2_ERC_721/csv_daywise"

# Create the output directory if it does not exist
if not os.path.exists(output_directory_daywise):
    os.makedirs(output_directory_daywise)

# Iterate through all files in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".csv"):
        # Load the CSV file into a DataFrame
        file_path = os.path.join(input_directory, filename)
        df = pd.read_csv(file_path)

        # Convert the timestamp column to datetime to extract dates
        df['timestamp'] = pd.to_datetime(df['timestamp'], format="%Y--%m--%d %H:%M:%S")
        
        # Extract the date part of the timestamp
        df['date'] = df['timestamp'].dt.date

        # Group by the date and save each group to a separate file
        for date, group in df.groupby('date'):
            # Create a filename for each date
            date_str = date.strftime("%Y-%m-%d")
            output_file_name = f"{date_str}.csv"
            output_file_path = os.path.join(output_directory_daywise, output_file_name)
            
            # Save the group DataFrame to a CSV file
            group.drop(columns='date', inplace=True)  # Drop the 'date' column if no longer needed
            group.to_csv(output_file_path, index=False)

            print(f"Saved data for {date_str} to {output_file_path}")
```
