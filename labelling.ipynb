```python
import os
import pandas as pd
import json

# Paths to the hash files
hashdict_file = os.path.join(r"C:\Mini Project 2", "hashdict.json")
hashdict_1_file = os.path.join(r"C:\Mini Project 2", "hashdict_1.json")
processed_files_file = os.path.join(r"C:\Mini Project 2", "processed_files.json")

# Load hash mappings from files if they exist and are valid
def load_json(file_path, default_value):
    if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
        try:
            with open(file_path, 'r') as file:
                return json.load(file)
        except json.JSONDecodeError:
            print(f"Error reading {file_path}. Initializing as empty.")
            return default_value
    else:
        return default_value

hashdict = load_json(hashdict_file, {})
id = max(hashdict.values()) + 1 if hashdict else 1

hashdict_1 = load_json(hashdict_1_file, {})
id_1 = max(hashdict_1.values()) + 1 if hashdict_1 else 1

processed_files = load_json(processed_files_file, [])

# Base directory path
base = r"C:\Mini Project 2\csv_daywisee"

# Output directory path
output = r"C:\Mini Project 2\labelling_outt"

directory = base

# Counter for processed files
processed_count = 0

# Iterate over all files in the directory
for filename in sorted(os.listdir(directory)):
    if filename in processed_files:
        print(f"Skipping already processed file: {filename}")
        continue

    f = os.path.join(directory, filename)

    # Read a CSV file to a pandas DataFrame
    df = pd.read_csv(f)

    # Two new columns for from and to label
    df['fromLabel'] = ''
    df['toLabel'] = ''
    df['tokenAddressLabel'] = ''

    # Iterate over all the rows in the DataFrame
    for i in range(len(df['from'])):
        if df.loc[i, 'from'] in hashdict:
            df.loc[i, 'fromLabel'] = hashdict[df.loc[i, 'from']]
        else:
            hashdict[df.loc[i, 'from']] = id
            df.loc[i, 'fromLabel'] = hashdict[df.loc[i, 'from']]
            id += 1

        if df.loc[i, 'to'] in hashdict:
            df.loc[i, 'toLabel'] = hashdict[df.loc[i, 'to']]
        else:
            hashdict[df.loc[i, 'to']] = id
            df.loc[i, 'toLabel'] = hashdict[df.loc[i, 'to']]
            id += 1

    for i in range(len(df['tokenAddressLabel'])):
        if df.loc[i, 'tokenAddress'] in hashdict_1:
            df.loc[i, 'tokenAddressLabel'] = hashdict_1[df.loc[i, 'tokenAddress']]
        else:
            hashdict_1[df.loc[i, 'tokenAddress']] = id_1
            df.loc[i, 'tokenAddressLabel'] = hashdict_1[df.loc[i, 'tokenAddress']]
            id_1 += 1

    # Extract only these columns
    df2 = df[['diffTimeStamp', 'fromLabel', 'toLabel', 'tokenAddressLabel']]
    new_directory = output
    f = os.path.join(new_directory, filename)
    print("Updated ", f)

    # Save to new CSV file
    df2.to_csv(f, index=False)

    # Add the filename to the list of processed files
    processed_files.append(filename)

    # Increment the counter for processed files
    processed_count += 1

    # Save the updated hash dictionaries and processed files to files
    with open(hashdict_file, 'w') as file:
        json.dump(hashdict, file)

    with open(hashdict_1_file, 'w') as file:
        json.dump(hashdict_1, file)

    with open(processed_files_file, 'w') as file:
        json.dump(processed_files, file)

print(f"Total files processed during this run: {processed_count}")
print(f"Total files processed overall: {len(processed_files)}")
# 2024--01--09.csv
```
