```python
import csv
from collections import defaultdict

# Input file (your CSV data)
input_file = r"E:\Major_Neeharika\4_hyperedges_data.csv"

# Dictionary to store token-wise node sets and their occurrence dates
token_nodes_date_map = defaultdict(lambda: defaultdict(list))

# Read the CSV file and collect tokens, node sets, and corresponding dates
with open(input_file, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        token = row['Token']
        date = row['Date']
        nodes = tuple(sorted(eval(row['Nodes'])))  # Convert the nodes string to a sorted tuple to ensure exact matching
        
        # Store the date where this token had this exact set of nodes
        token_nodes_date_map[token][nodes].append(date)

# Find tokens and node sets that have been repeated on different days
repeated_tokens_data = []
for token, nodes_dates in token_nodes_date_map.items():
    for nodes, dates in nodes_dates.items():
        if len(dates) > 1:  # If the same set of nodes appeared on multiple dates
            repeated_tokens_data.append({
                'Token': token,
                'Nodes': nodes,
                'Dates': dates
            })

# Output the repeated tokens and their occurrence dates
output_file = r"D:\SEM_8\Major Project\4_hyperedges_repeated_tokens.csv"

# Write the result to a CSV file
with open(output_file, 'w', newline='') as csvfile:
    fieldnames = ['Token', 'Nodes', 'Dates']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write the header
    writer.writeheader()
    
    # Write the repeated tokens, nodes, and their occurrence dates
    for entry in repeated_tokens_data:
        writer.writerow({
            'Token': entry['Token'],
            'Nodes': entry['Nodes'],
            'Dates': ", ".join(entry['Dates'])  # Convert the list of dates to a comma-separated string
        })

print("Repeated tokens and nodes with dates have been successfully saved to CSV.")
```

## Temporal Distance

```python
import pandas as pd

# Input file path
input_file = r"D:\SEM_8\Major Project\smart_filtered_4_output.csv"

# Read CSV file
df = pd.read_csv(input_file)

# Ensure Dates column is properly processed by stripping spaces and converting to datetime
df['Dates'] = df['Dates'].apply(lambda x: sorted(pd.to_datetime([date.strip() for date in x.split(',')])))

# Function to compute temporal distances
def calculate_temporal_distances(dates):
    return [abs((dates[i] - dates[i-1]).days) for i in range(1, len(dates))]

# Compute temporal distances
df['Temporal_Distance'] = df['Dates'].apply(lambda x: calculate_temporal_distances(x))

# Convert dates to "YYYY-MM-DD" format (remove timestamps)
df['Dates'] = df['Dates'].apply(lambda x: str([date.strftime("%Y-%m-%d") for date in x]))

# Convert the list of temporal distances to a string format
df['Temporal_Distance'] = df['Temporal_Distance'].apply(lambda x: str(x))

# Save to output CSV file
output_file = r"D:\SEM_8\Major Project\temporal_distances_4_output.csv"
df.to_csv(output_file, index=False)

print(f"Temporal distances saved to: {output_file}")
```

## Bridges

```python
import os
import pandas as pd
import networkx as nx
from ast import literal_eval

# File paths
input_file = r"C:\Users\Neeharika Telu\Dropbox\Mini_clean\4_hyperedges_data.csv"
output_file = r"D:\SEM 8\Major Project\4_bridge_output.csv"

# Ensure output directory exists
output_dir = os.path.dirname(output_file)
os.makedirs(output_dir, exist_ok=True)

# Read CSV file
df = pd.read_csv(input_file, delimiter=",", dtype=str, skipinitialspace=True, engine="python")

# Convert Nodes column from string to list
df["Nodes"] = df["Nodes"].apply(literal_eval)

# Create a dictionary to store bridge nodes and their hyperedges
bridge_nodes = {}

# Create a NetworkX graph to analyze topological distance
G = nx.Graph()

# Build graph edges
for _, row in df.iterrows():
    hyperedge = row["Nodes"]
    token = row["Token"]

    # Add hyperedges to the graph
    for i in range(len(hyperedge)):
        for j in range(i + 1, len(hyperedge)):
            G.add_edge(hyperedge[i], hyperedge[j], token=token)

    # Identify bridge nodes (nodes appearing in multiple hyperedges)
    for node in hyperedge:
        if node not in bridge_nodes:
            bridge_nodes[node] = {"token": token, "hyperedges": []}
        bridge_nodes[node]["hyperedges"].append(hyperedge)

# Find shortest path distances (topological distance)
topological_distances = {}
for _, row in df.iterrows():
    hyperedge = row["Nodes"]
    distances = []
    for i in range(len(hyperedge)):
        for j in range(i + 1, len(hyperedge)):
            if nx.has_path(G, hyperedge[i], hyperedge[j]):
                distances.append(nx.shortest_path_length(G, hyperedge[i], hyperedge[j]))
    avg_distance = sum(distances) / len(distances) if distances else 0
    topological_distances[str(hyperedge)] = avg_distance

# Convert bridge nodes to output format
output_data = []
for node, details in bridge_nodes.items():
    output_data.append({
        "Bridge Node": node,
        "Token": details["token"],
        "Hyperedges": details["hyperedges"]
    })

# Convert output to DataFrame and save to CSV
output_df = pd.DataFrame(output_data)
output_df.to_csv(output_file, index=False)

print(f"Processed output saved to {output_file}")
```

```python
import pandas as pd
import ast
from collections import defaultdict

# Load CSV file
def process_hyperedges(file_path):
    df = pd.read_csv(file_path, header=None, names=["id", "token", "hyperedges"], dtype={"id": str, "token": str, "hyperedges": str})
    
    # Function to safely evaluate hyperedges
    def safe_eval(val):
        try:
            return ast.literal_eval(val) if isinstance(val, str) and val.startswith("[[") else []
        except (ValueError, SyntaxError):
            return []
    
    # Convert hyperedges from string to list
    df["hyperedges"] = df["hyperedges"].apply(safe_eval)
    
    # Dictionary to count occurrences of nodes
    node_count = defaultdict(int)
    
    # Count how many hyperedges each node appears in
    for hyperedge_list in df["hyperedges"]:
        unique_nodes = set(node for hyperedge in hyperedge_list for node in hyperedge)
        for node in unique_nodes:
            node_count[node] += 1
    
    # Function to filter bridges based on connectivity
    def filter_bridges(hyperedge_list):
        unique_nodes = set(node for hyperedge in hyperedge_list for node in hyperedge)
        if len(unique_nodes) > 1:
            # Sort nodes by highest hyperedge count
            sorted_nodes = sorted(unique_nodes, key=lambda n: node_count[n], reverse=True)
            return sorted_nodes[0]  # Keep only the highest-degree node as bridge
        return list(unique_nodes)
    
    # Apply filtering
    df["filtered_bridges"] = df["hyperedges"].apply(filter_bridges)
    
    # Save cleaned file
    output_path = file_path.replace(".csv", "_cleaned.csv")
    df.to_csv(output_path, index=False)
    
    # Save output in a separate CSV file
    output_csv_path = file_path.replace(".csv", "_output.csv")
    df.to_csv(output_csv_path, index=False)
    
    print(f"Processed file saved as: {output_path}")
    print(f"Output saved in: {output_csv_path}")

# Example usage
file_path = r"D:\SEM 8\Major Project\4_bridge_output.csv"
process_hyperedges(file_path)
```

### Frequency of the Bridge Nodes

```python
import pandas as pd
import ast
from collections import defaultdict

# Load CSV file
def process_hyperedges(file_path):
    df = pd.read_csv(file_path, header=None, names=["id", "token", "hyperedges"], dtype={"id": str, "token": str, "hyperedges": str})
    
    # Function to safely evaluate hyperedges
    def safe_eval(val):
        try:
            return ast.literal_eval(val) if isinstance(val, str) and val.startswith("[[") else []
        except (ValueError, SyntaxError):
            return []
    
    # Convert hyperedges from string to list
    df["hyperedges"] = df["hyperedges"].apply(safe_eval)
    
    # Compute frequency of bridge nodes
    def compute_bridge_frequency(hyperedge_list):
        unique_hyperedges = set(tuple(sorted(h)) for h in hyperedge_list)  # Ensure uniqueness
        bridge_frequencies = defaultdict(int)
        for hyperedge in unique_hyperedges:
            for node in hyperedge:
                bridge_frequencies[node] += 1  # Count unique hyperedges a node appears in
        return dict(bridge_frequencies)
    
    # Apply computation and store results
    df["bridge_frequencies"] = df["hyperedges"].apply(compute_bridge_frequency)
    
    # Save cleaned file
    output_path = file_path.replace(".csv", "_cleaned.csv")
    df.to_csv(output_path, index=False)
    
    # Save output in a separate CSV file
    output_csv_path = file_path.replace(".csv", "_output.csv")
    df.to_csv(output_csv_path, index=False)
    
    print(f"Processed file saved as: {output_path}")
    print(f"Output saved in: {output_csv_path}")

# Example usage
file_path = r"D:\SEM 8\Major Project\4_bridge_output.csv"
process_hyperedges(file_path)
```

### 5 Hyperedge

```python
import os
import json
import csv

# Define the input folder path and output file
input_folder = r"E:\Major_Neeharika\hypergraph_everyday_Null_clean"
output_file = "5_hyperedges_data.csv"

# Open the output CSV file for writing
with open(output_file, mode='w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Date", "Token", "Nodes"])
    
    # Iterate through each file in the folder
    for filename in sorted(os.listdir(input_folder)):
        if filename.endswith(".json"):  # Ensure it's a JSON file
            file_path = os.path.join(input_folder, filename)
            
            # Extract date from filename (assuming format: hypergraph_YYYY-MM-DD.json)
            date = filename.replace("hypergraph_", "").replace(".json", "")
            
            try:
                # Read JSON file
                with open(file_path, 'r') as file:
                    data = json.load(file)
                
                # Extract 5-hyperedges (tokens with exactly 5 nodes)
                for token, nodes in data.items():
                    if isinstance(nodes, list) and len(nodes) == 5:
                        writer.writerow([date, token, json.dumps(nodes)])
            except Exception as e:
                print(f"Error processing file {filename}: {e}")

print(f"Extraction completed. Data saved in {output_file}")
```
