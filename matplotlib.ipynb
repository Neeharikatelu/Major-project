```python
G = nx.MultiDiGraph()
```

```python
for row in df.iterrows():
  if row[1].fromLabel not in G:
      G.add_node(row[1].fromLabel)
  if row[1].toLabel not in G:
      G.add_node(row[1].toLabel)
  G.add_edge(row[1].fromLabel,row[1].toLabel)
```

```python
degree_sequence = sorted((d for n, d in G.out_degree()), reverse=True)
N = len(G.nodes())
degreelist = []
for x in degree_sequence:
  if x != 0:
    degreelist.append(x)


degreelist, frequencies = np.unique(degree_sequence, return_counts=True)
probabilities = [i/N for i in frequencies]
```

```python
plt.scatter(degreelist, probabilities)
plt.xlabel("Out-Degree(k)", fontweight = "bold", fontsize = 15)
plt.ylabel("P(k)", fontweight = "bold", fontsize = 15)
plt.xscale('log')
plt.yscale('log')
plt.rcParams['axes.linewidth'] = 3
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.show()
#plt.savefig('indegreedistribution.png',bbox_inches = 'tight',dpi=300)
```
